```python
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from sklearn import datasets
```

# Barcharts

```python
num_bins = 3
bar1=np.random.randint(0,100,num_bins)
bar2=np.random.randint(0,100,num_bins)
width = 0.25
indices = np.arange(num_bins)
plt.bar(indices,bar1,width)
plt.bar(indices+width,bar2,width,label='bar2')
```

```python
plt.xticks(indices,('A','B','C'))
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

# Histograms

```python
mean = 5.0
deviation = 2.0
vals = mean + deviation*np.random.randn(1000)
num_bins = 100
plt.hist(vals,num_bins)
```

```python
plt.xlabel('X')
plt.title('Title')
plt.grid(True)
plt.show()
```

# 3D Line Plots

```python
phi = np.linspace(-6*np.pi,6*np.pi,100)
z = np.linspace(-4,4,100)
x = np.sin(phi)
y = np.cos(phi)
```

```python
plt.show()
```

# 3D Surface Plots

```python
X,Y = np.meshgrid(np.arange(-10,10),np.arange(-10,10))
Z = X**2 + Y**2
G_x,G_y = np.gradient(Z)
G = np.sqrt(G_x**2+G_y**2)
G = G / G.max()
```

```python
fig =plt.figure()
axes = fig.gca(projection = '3d')
axes.plot_surface(X,Y,Z, facecolors = cm.inferno(G))
```

```python
plt.show()
```
